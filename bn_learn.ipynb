```python

import pandas as pd
import numpy as np
np.random.seed(42)
import random
random.seed(42)
import bnlearn as bn
# Load titanic dataset containing mixed variables

#df_raw = bn.import_example(data='titanic')

# Pre-processing of the input dataset
#dfhot, dfnum = bn.df2onehot(df_raw)

# Structure learning


csv_file = 'data/NonCausalVsCausal_CausalPlan.xlsx' 
        
data = pd.read_excel(csv_file)
data.drop(columns=data.columns[0], axis=1, inplace=True)

model_hc_bic  = bn.structure_learning.fit(data, methodtype='hc', scoretype='bic')
model_cs_bic = bn.independence_test(model_hc_bic, data)


model_hc_k2   = bn.structure_learning.fit(data, methodtype='hc', scoretype='k2')
model_cs_bic = bn.independence_test(model_hc_k2, data)

model_hc_bdeu = bn.structure_learning.fit(data, methodtype='hc', scoretype='bdeu')
model_cs_bic = bn.independence_test(model_hc_bdeu, data)

model_hc_aic= bn.structure_learning.fit(data, methodtype='hc', scoretype='aic')
model_cs_bic = bn.independence_test(model_hc_aic, data)


model_ex_bic  = bn.structure_learning.fit(data, methodtype='ex', scoretype='bic')
model_cs_bic = bn.independence_test(model_ex_bic, data)

model_ex_k2   = bn.structure_learning.fit(data, methodtype='ex', scoretype='k2')
model_cs_bic = bn.independence_test(model_ex_k2, data)

model_ex_bdeu = bn.structure_learning.fit(data, methodtype='ex', scoretype='bdeu')
model_cs_bic = bn.independence_test(model_ex_bdeu, data)

model_cs_bic = bn.structure_learning.fit(data, methodtype='cs')
# Compute edge strength using chi-square independence test
model_cs_bic = bn.independence_test(model_cs_bic, data)
# Plot based on structure learning of sampled data

```

```python

print('methodtype; ', 'hc', ' | scoretype; ',  'bic')
G = bn.plot(model_hc_bic)

print('methodtype; ', 'hc', ' | scoretype; ',  'k2')
G = bn.plot(model_hc_k2)

print('methodtype; ', 'hc', ' | scoretype; ',  'aic')
G = bn.plot(model_hc_aic)

print('methodtype; ', 'hc', ' | scoretype; ',  'bdeu')
G = bn.plot(model_hc_bdeu)


print('methodtype; ', 'ex', ' | scoretype; ',  'bic')
G = bn.plot(model_ex_bic)

print('methodtype; ', 'ex', ' | scoretype; ',  'k2')
G = bn.plot(model_ex_k2)

print('methodtype; ', 'ex', ' | scoretype; ',  'bdeu')
G = bn.plot(model_ex_bdeu)


print('methodtype; ', 'cs', ' | scoretype; ',  'chi2')
G = bn.plot(model_cs_bic)

```
